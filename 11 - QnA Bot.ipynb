インスタントメッセージ、ソーシャルメディア、電子メール、その他のオンライン技術を使って他の人とコミュニケーションをとる頻度を考えてみてください。多くの人にとって、それは日常的な連絡手段となっています。仕事で質問があるときには、モバイルデバイスで使えるチャットメッセージを使って同僚に連絡を取るかもしれません。

ボットは、このようなチャネルを使ってコミュニケーションを行うAIエージェントで、ソフトウェアサービスとの自然な会話を可能にします。

## QnA Maker のナレッジベースを作成する

カスタマーサポートのシナリオでは、ウェブサイトのチャットウィンドウ、電子メール、または音声インターフェイスを通じて、よくある質問を解釈して回答するボットを作成するのが一般的です。ボットのインターフェースの基礎となるのは、質問と適切な回答のナレッジベースであり、ボットは適切な回答を検索することができます。

QnA Makerサービスは、AzureのCognitive Serviceで、質問と回答のペアを入力したり、既存のドキュメントやWebページからナレッジベースを素早く作成することができます。また、内蔵されている自然言語処理機能を使って、質問を解釈し、適切な回答を見つけることができます。

1. 別のブラウザタブを開き、QnA Maker ポータル（https://qnamaker.ai）にアクセスします。Azureサブスクリプションに関連付けられたMicrosoftアカウントを使用してサインインします。
2. QnA Maker ポータルで、**Create a knowledge base** を選択します。
3. 以前にQnAサービスリソースを作成していない場合は、**Create a QnA service** をクリックします。Azureポータルが別のタブで開かれるので、サブスクリプションにQnA Maker サービスを作成することができます。以下の設定を行なってください。
   
   - **サブスクリプション**: *ご自身のサブスクリプション*
   - **リソースグループ**: *既存のリソースグループを選択するか、ユニークな名前で新しいリソースグループを作成します。*
   - **名前**: *ご自身のQnAリソース固有の名前*
   - **価格レベル**: F0
   - **Azure Search の場所**: *好みの場所を選択*
   - **Azure Search の価格レベル**: F
   - **アプリ名**: **名前** と同じ _(名前の後に ".azurewebsites.net" が自動的に追加されます。)_
   - **Web サイトの場所**: **Azure Search の場所** と同じ
   - **App insights**: 無効

> **注意** すでにFree階層の **QnA Maker** または**Azure Search** リソースをプロビジョニングしている場合、クォータでは別のリソースを作成できない場合があります。その場合は、**F0** / **F** 以外の階層を選択してください。

4. QnAサービスと関連リソースの展開がAzureポータルで完了するのを待ちます。
5. QnA Makerポータルタブに戻り、**STEP 2** のセクションで **Refresh** をクリックして、利用可能なQnAサービスリソースのリストを更新します。
6. 以下のオプションを選択して、QnAサービスとKB(ナレッジベース)を接続します。
  - **Microsoft Azure Directory ID**: *サブスクリプションのAzureディレクトリID*
  - **Azure subscription name**: *ご自身のサブスクリプション*
  - **Azure QnA service**: *前のステップで作成したQnAサービスリソース*
  - **Language**: English

### (!) チェックイン
ロールにアクションを実行する権限がないというメッセージが表示された場合は、QnA Makerポータルのブラウザページを更新してください。

7. **STEP 3** のセクションでは、**Margie's Travel KB** という名前を入力します。
8. **STEP 4** のセクションでは、**URL**ボックスに *https://github.com/MicrosoftLearning/mslearn-ai900/raw/main/data/qna_bot/margies_faq.docx* と入力し、**+ Add URL** をクリックします。**Chit-chat**の選択項目は**Professional** を選択します。
9. **STEP 5** のセクションでは、**Create your KB** をクリックします。
10. ナレッジベースが作成されるまで、1分ほどお待ちください。その後、FAQドキュメントからインポートされた質問と回答、およびプロの雑談用の事前定義された回答を確認します。


## ナレッジベースを編集する

ナレッジベースは、FAQドキュメントの詳細と、あらかじめ定義されたいくつかの回答に基づいています。これらを補完するために、カスタムの質問と回答のペアを追加することができます。

1. **+ Add QnA pair** をクリックします。
2. **Question** のボックスに、`Hello` を入力します。次に **+ Add alternative phrasing** をクリックし `Hi` と入力します。
3. **Answer** のボックスに、`Hello` を入力します。

## ナレッジベースのトレーニングとテスト

これでナレッジベースができたので、QnA Makerポータルでテストすることができます。

1. ページの右上にある「**Save and train**」をクリックして、ナレッジベースをトレーニングします。このボタンを表示するには、ウィンドウのサイズを変更する必要があるかもしれません。
2. トレーニングが完了したら、**&larr; Test** をクリックして、テストペインを開きます。
3. テスト画面の一番下に、メッセージ *Hi* と入力します。**Hello** という応答が表示されるはずです。
4. テスト画面の一番下に、*I want to book a flight* というメッセージを入力します。FAQから適切な回答が返されるはずです。
5. ナレッジベースのテストが終了したら、**&rarr; Test** をクリックしてテストペインを閉じます。


## ナレッジベースのためのボットを作成する

ナレッジベースは、クライアントアプリケーションがある種のユーザーインターフェースを通じて質問に答えるために使用できるバックエンドサービスを提供します。一般的に、これらのクライアントアプリケーションはボットです。ボットがナレッジベースを利用できるようにするには、HTTPでアクセスできるサービスとしてナレッジベースを公開する必要があります。その後、Azure Bot Serviceを使用して、ナレッジベースを使用してユーザーの質問に答えるボットを作成し、ホストすることができます。

1. QnA Makeのページの上部にある、**Publish**をクリックします。次に、**Margies Travel KB** のページで、**Publish** をクリックします。
2. サービスがデプロイされたら、**Create Bot** をクリックします。これにより、新しいブラウザタブでAzureポータルが開き、AzureサブスクリプションにWeb App Botを作成することができます。
3. Azureポータルで、以下の設定でWeb App Botを作成します（これらのほとんどは事前に入力されています）。
  - **ボット ハンドル**: *ボット固有の名前をつけます*
  - **サブスクリプション**: *ご自身のサブスクリプション*
  - **リソースグループ**: *既存のリソースグループを選択するか、ユニークな名前で新しいリソースグループを作成します。*
  - **場所**: *先に作成した QnA Maker service と同じ場所*.
  - **価格レベル**: F0
  - **アプリ名**: * **ボット ハンドル** と同じ名前  _(*.azurewebsites.net* が自動的に付加されます)_
  - **SDK 言語**: *C# か Node.js のどちらかを選択します*
  - **QnA の認証キー**: *QnAナレッジベースの認証キーが自動的に設定されます。*
  - **App Service プラン/場所**: *適切なプランと場所が自動的に設定されます。*
  - **Application Insights**: オフ
  - **アプリID とパスワードの自動作成**: アプリIDとパスワードの自動作成
4. ボットが作成されるのを待ちます（右上のベルのような通知アイコンは、待っている間にアニメーションします）。その後、デプロイメントが完了した通知で、**リソースに移動** をクリックします（または、ホームページで **リソースグループ** をクリックし、ウェブアプリのボットを作成したリソースグループを開いてクリックします）。
5. ボットの左側のペインで **設定** を探し、**Web チャットでテスト** をクリックし、ボットが **Hello and welcome!** のメッセージを表示するまで待ちます（初期化に数秒かかる場合があります）。
6. テストチャットインターフェースを使用して、ボットがナレッジベースからの質問に期待通りに答えることを確認します。例えば、*I need to cancel my hotel* を送信してみてください。
